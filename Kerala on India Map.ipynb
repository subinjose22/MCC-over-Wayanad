```python
import geopandas as gpd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Load shapefiles
india = gpd.read_file('/Users/subin/Downloads/Shape_files/India/India_Country_Boundary.shp')
kerala = gpd.read_file("/Users/subin/Downloads/Shape_files/kerala/shapefiles/state.shp")

# Ensure CRS is WGS84 (longitude-latitude)
india = india.to_crs(epsg=4326)
kerala = kerala.to_crs(epsg=4326)

# Set up Cartopy map with PlateCarree projection (for longitude-latitude)
fig, ax = plt.subplots(figsize=(12, 12), subplot_kw={'projection': ccrs.PlateCarree()})

# Add land, ocean, borders for context (optional)
ax.add_feature(cfeature.LAND, facecolor='white')
#ax.add_feature(cfeature.BORDERS, linewidth=0.5)
#ax.coastlines(resolution='10m', linewidth=0.5)

# Plot India boundary

india.boundary.plot(ax=ax, edgecolor='black', linewidth=2, transform=ccrs.PlateCarree())

# Plot Kerala with fill
kerala.plot(ax=ax, color='brown', edgecolor='black', linewidth=0.8, transform=ccrs.PlateCarree())

# Set extent (optional: zoom to India)
ax.set_extent([68, 98, 6, 38], crs=ccrs.PlateCarree())

# Add gridlines and labels
gl = ax.gridlines(draw_labels=True, linestyle='--', linewidth=1, color='gray')
gl.top_labels = False
gl.right_labels = False
gl.xlabel_style = {'size': 24}
gl.ylabel_style = {'size': 24}

# Title
#plt.title('Kerala on India Map with Geolocations (Cartopy)', fontsize=16)
plt.savefig('Kerala on India Map', dpi=300)
plt.show()
```
